# Data Cleaning & Manipulation

## 1. Handle Missing Data

```python
# Check for missing values
df.isnull().sum()
```

### Insights
- No missing values were found in the dataset.
- Therefore, no imputation or removal was required.

## 2. Remove Duplicates

```python
# Check for duplicate rows
df.duplicated().sum()
```

### Insights
- No duplicate records were found.
- No rows were removed.

## 3. Fix Inconsistencies in Categorical Data

```python
# Standardize categorical columns
cat_cols = ['Day of the week','Traffic Situation']

for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()

df[cat_cols].head()
```

### Insights
- Case and whitespace inconsistencies were fixed.
- Categories are now standardized.

## 4. Convert Data Types

```python
# Convert Date and Time to datetime
df["Time"] = pd.to_datetime(df["Time"])
# convert Object to category
df["Traffic Situation"]=df["Traffic Situation"].astype('category')
```

### Insights
- Time is converted to datetime.
- Traffic Situation is converted to category dtype

## Data visualization

## 1.Univariate visualization

```python
#Numerical
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
#  Histogram
ax[0][0].hist(df["CarCount"], bins=20)
ax[0][0].set_title("Distribution of CarCount")
ax[0][0].set_xlabel("CarCount")
ax[0][0].set_ylabel("Frequency")
#  KDE Plot
df["BikeCount"].plot(kind='kde', ax=ax[0][1],color="red")
ax[0][1].set_title("Distribution of BikeCount")
ax[0][1].set_xlabel("BikeCount")
ax[0][1].set_ylabel("Density")

# categorical
sns.countplot(data=df, x="Traffic Situation", ax=ax[1][0],color="Black")
ax[1][0].set_title("Count Plot of Traffic Situation")
ax[1][0].set_xlabel("Traffic Situation")
ax[1][0].set_ylabel("Count")
df["Traffic Situation"].value_counts().plot(kind="pie",ax=ax[1][1])
ax[1][1].set_title("Proportion of Traffic Situation")
plt.show()
```


## Insights

- **CarCount:**  
  The histogram shows that car counts are concentrated within a specific range, indicating that most observations experience moderate car traffic.

- **BikeCount:**  
  The density plot shows a clear peak, suggesting that bike traffic is most common around certain values and varies smoothly across observations.


## Insights

- The count plot shows that the distribution of traffic situations is not uniform, with some categories occurring more frequently than others.

- The bar chart confirms the same frequency pattern using a different visualization approach, making it easy to compare category counts.

- The pie chart highlights the proportion of each traffic situation, clearly showing which category dominates the dataset.


## Bivariate visualization

```python
## Numerical-Numerical 
fig, ax = plt.subplots(3, 2, figsize=(25, 20))
sns.scatterplot(data=df,x="CarCount",y="Total",ax=ax[0, 0])
ax[0, 0].set_title("CarCount vs Total")
sns.lineplot(data=df, x="BikeCount", y="CarCount", ax=ax[0][1])
ax[0, 1].set_title("BikeCount vs CarCount")

##Numerical-categorical
sns.barplot(data=df, x="Day of the week", y="BusCount", ax=ax[1, 0], color="green")
ax[1, 0].set_title("Bar plot")
sns.stripplot(data=df,x="Traffic Situation",y="TruckCount",ax=ax[1, 1])
ax[1, 1].set_title("Average TruckCount by Traffic Situation")

#Categorical-categorical
sns.countplot(data=df, x="Day of the week", hue="Traffic Situation", ax=ax[2][0])
ax[2][0].set_title("Grouped Count Plot")
df1 = pd.crosstab(df["Day of the week"], df["Traffic Situation"])
df1.plot(kind="bar", stacked=True, ax=ax[2][1])
ax[2][1].set_title("Stacked bar")
plt.show()
```

## Insights

- **CarCount vs Total:**  
  The scatter plot shows a clear upward pattern, indicating that as the number of cars increases, the total traffic volume also increases. This suggests that cars are the main contributors to overall traffic.

- **BikeCount vs CarCount:**  
  The points show a moderate positive relationship, meaning that locations or times with more bikes generally also have more cars. However, the spread of points indicates that this relationship is not very strong.

  **BusCount vs Day of the week:**  
  The bar plot indicates that bus traffic remains relatively low and consistent across most days, showing limited variation in bus usage.

- **TruckCount vs Traffic Situation:**  
  This plot shows differences in average truck counts across traffic situations, indicating that higher truck counts are generally associated with heavier traffic conditions.
- **Grouped Count Plot (Day of the week vs Traffic Situation):**  
  The plot shows that traffic situations vary across different days of the week, with certain days experiencing higher levels of heavy or medium traffic.

- **Stacked Bar Plot:**  
  The stacked bars highlight both the total number of observations per day and the distribution of traffic situations within each day, indicating that some days have a higher proportion of heavy traffic.
